```python
import urllib.request
import urllib.parse
import re
import json

search = '부자'
category = 'book'
display = 20
start = 1
sort = 'sim'

baseURL = 'https://openapi.naver.com/v1/search/'
node = f'{category}.json'

param = f'?query={urllib.parse.quote(search)}'
param += f'&display={display}'
param += f'&start={start}'
param += f'&sort={sort}'

URL = baseURL + node + param
```

```python
request = urllib.request.Request(URL)
request.add_header("X-Naver-Client-Id", "XFPmh3jDdh_P0Q76kvAG")
request.add_header("X-Naver-Client-Secret", "t1hqy18uNp")

response = urllib.request.urlopen(request)
rescode = response.getcode()

if rescode == 200:
    response_body = response.read()
    res_data = response_body.decode('utf-8')

    # 문자열을 JSON 객체로 변환
    try:
        json_data = json.loads(res_data)
    except json.JSONDecodeError as json_ex:
        print(f"JSON 디코딩 에러: {json_ex}")
else:
    print("Error Code:", rescode)
```
